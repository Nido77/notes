# 1. 举个栗子

```python
import torch
from torchvision.models import resnet18, ResNet18_Weights
model = resnet18(weights=ResNet18_Weights.DEFAULT)
data = torch.rand(1, 3, 64, 64)
labels = torch.rand(1, 1000)
```

```python
prediction = model(data) # forward pass
print(prediction)
```

```python
loss = (prediction - labels).sum()
loss.backward() # backward pass
# 自动计算梯度并存储与varName.grad中
```

```python
# 优化器, 这里是用的是SGD, lr = 0.01, momentum = 0.9
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
```

```python
optim.step() #gradient descente
```

# 2. Autograd的一些细节

```python
import torch

a = torch.tensor([2., 3.], requires_grad=True)      # 默认为false
b = torch.tensor([6., 4.], requires_grad=True)
```

$$
Q = 3a^3 - b^2
$$

```python
Q = 3*a**3 - b**2
print(Q)
```

```python
# 对Q进行求导,因为Q是一个矩阵, 所以要进行求和才能求导
Q.sum().backward()

# 这种写法和上面是一样的
# external_grad = torch.tensor([1., 1.]) 对Q进行点乘
# Q.backward(gradient=external_grad)
```

```python
# 检查梯度的正确性
print(9*a**2 == a.grad)
print(-2*b == b.grad)
```
